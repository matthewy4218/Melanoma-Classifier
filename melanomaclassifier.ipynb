```python
from fastai.vision.all import *
import kagglehub
from pathlib import Path
```

```python
# Download dataset using kagglehub
download_path = kagglehub.dataset_download("hasnainjaved/melanoma-skin-cancer-dataset-of-10000-images")

# The actual data is inside this subfolder:
data_path = Path(download_path) / "melanoma_cancer_dataset"
# Confirm structure
print((data_path / 'train').ls())  # should show ['benign', 'malignant']
```

```python
dls = ImageDataLoaders.from_folder(
    data_path / 'train',
    valid_pct=0.2,
    seed=42,
    item_tfms=Resize(224),
    batch_tfms=aug_transforms()
)
```

```python
learn = vision_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(3)
```

```python
# Just export with compatible pickle protocol
learn.export('melanoma_classifier.pkl')
```
